## Conditionally Adding Column

To add Column in our excelsheet {whoose value will depand on condition}, we'll import numpy module.

```python
import numpy as np
```

```python
df["Health Status"]=np.where(df["AGE"]==18,"Fit","Unfit")
```

```python
df.head()
```

> Adding Column with same value 

```python
df["Year"]=np.array(2023)
```

```python
df.head()
```

> Drop a Column Permanently

```python
df.drop(["Health Status"],axis = 1,inplace=True)
```

```python
df.head()
```

```python
HealthStatus=[]
```

```python
for i in df["AGE"]:
    if i<=20:
        HealthStatus.append("Very Good")
    elif i>20 and i<30:
        HealthStatus.append("Good")
    else:
        HealthStatus.append("Not Good")
        
```

```python
df.drop(["Health Status"],axis=1)
```

```python
df["Health Status"]=HealthStatus
```
