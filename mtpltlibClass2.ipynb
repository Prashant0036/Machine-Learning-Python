```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= [1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
#plt.xlabel('Day')
plt.ylabel('Temp.')
plt.plot(x,y)
plt.title('Weather')
plt.plot(x,y,color='orange', linewidth=5, linestyle='dotted')
```

> By ploting any data on graph, we can better visualize it and can make descisons easily.

```python
plt.plot(x, y, color='red',marker='D', linestyle='--',markersize=5)
```

```python
plt.plot(x, y, "*--b",alpha=0.5)      #alpha defines the opacity bw 0 to 1
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50, 51, 52,48,47,49,46]
min_t = [43,42,40,44, 33, 35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend (loc="upper right")
plt.grid()
```
