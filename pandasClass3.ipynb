## Grouping of Data using Groupby function.

```python
df.dtypes
```

> First of all convert "perc" column in float datatypes from string 

```python
df["perc"]=pd.to_numeric(df["perc"],errors="coerce")
```

```python
df.dtypes
```

```python
df.groupby("owner_ship")["perc"].mean() #or df.groupby("owner_ship")["perc"].agg("mean")
```

```python
df.groupby("owner_ship")["oi"].mean() #or df.groupby("owner_ship")["oi"].agg("mean")
```

```python
df.groupby(["name","owner_ship"])["perc"].agg(["max"])
```

```python
df["total"]=pd.to_numeric(df["total"],errors="coerce")
```

> If you've to perform operation on multiple column before group them.

```python
processedResult=df.groupby("owner_ship").agg({"perc":["mean"],"total":["max","min"]})
```

```python
processedResult
```

```python
processedResult.columns=["PercMean","totalMax","totalMin"]
```

```python
processedResult
```

> Convert any dataframe into a csv file

```python
processedResult.to_csv("processedResult1.csv")
```
