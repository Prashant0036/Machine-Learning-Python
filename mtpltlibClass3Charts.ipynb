```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Bar Chart

```python
company=["GOOGL", "AMZN", "MSFT", "FB" ]
revenue=[90,136,89,27]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue,width=0.3)
plt.xticks(xpos,company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Tech Stocks")
plt.legend("R")
```

```python
company=["GOOGL", "AMZN", "MSFT", "FB" ]
revenue=[90,136,89,27]
profit=[40,80,35,10]
```

```python
plt.bar(xpos,revenue,width=0.3) #Both value should be numeric,
plt.xticks(xpos,company)
plt.bar(xpos+0.5,profit,width=0.3)
plt.legend (loc="upper right")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Tech Stocks")
```

### Horizontal Bar Chart

```python
plt.yticks (xpos, company)
plt.title("US Tech Stocks")
plt.barh(xpos-0.2, revenue, label="Revenue")
plt.barh(xpos+0.2, profit, label="Profit")
plt.legend()
```

## Histogram

Histogram draws 1D array on x axis and the frequecies on Y axis.
A histogram, on the other hand, is a graph that shows the distribution of numerical data. 
It is a type of bar chart that shows the frequency or number of observations within different numerical ranges, called bins.

```python
blood_sugar= [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,width=5)
```

```python
plt.hist(blood_sugar,bins=[90,115,120,150],rwidth=0.8,color='orange')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins= [80,100,125,150], rwidth=0.95, color=["green","orange"],
label=['men', 'women'])
plt.legend()
```

```python
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins= [80,100,125,150], rwidth=0.95, color=["green","orange"],
label=['men', 'women'],orientation='horizontal')
plt.legend()
```

## Pie Chart

```python
exp_vals = [1400, 600, 300, 410,250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis ("equal")
plt.pie(exp_vals, labels=exp_labels,radius=1.5, autopct='%0.2f%%', shadow=True, explode= [0, 0,0.2,0,0],startangle=30)
plt.show()
```
